### Language Models - N-Gram Approach

#### 1. Introduction
- A Language Model (LM) is a probablistic system that assigns a probability to a sequence of words.
- Why Language Models:
1. Predict next word (autocomplete)
2. Understands sentence flow 
3. Generate human-like text 



#### 2. Mathematical Foundation
- For a sentence P(w1, w2, w3...., wn)
- Using chain rule: P(w1​,w2​,...,wn​)=P(w1​)×P(w2​∣w1​)×P(w3​∣w1​,w2​)×...

Problem (Why N-Gram?)

Full dependency is:
1. Computationally expensive
2. Requires huge data 

Solution: Markov Assumption: P(wn​∣w1​,...,wn−1​)≈P(wn​∣last N−1 words)

#### 3. What is N-Gram
An N-gram is a sequence of N consecutive words
- Unigram (No context)  = I, Love
- Bigram (1 word context) = I love
- Trigram (2 words context) = I love NLP

#### 4. Complete Algorithm
1. Input text corpus
2. Convert text to lowercase
3. Tokenize text into words
4. Generate N-grams
5. Count frequencies of N-grams
6. Count frequencies of (N-1)-grams
7. Calculate probabilities:

        P(wi | wi-1) = Count(wi-1, wi) / Count(wi-1)
8. Store probabilities
9. Generate sentences using probabilities
10. Evaluate using perplexity

```python
def preprocess(corpus):
    tokens = []
    for sentence in corpus:
        sentence= sentence.lower()
        words = sentence.split()
        tokens.append(words)
    return tokens

tokens = preprocess(corpus)
print(tokens)
```

##### Step 3: Generate N-Grams
- Why? To capture words relationships
- How? 
    - ["i","love", "nlp"]
    - Unigram → ["i", "love", "nlp"]
    - Bigram → [("i","love"), ("love","nlp")]
    - Trigram → [("i","love","nlp")]

```python
def generate_bigrams(tokens):
    bigrams = []
    for sentence in tokens:
        for i in range (len(sentence)-1):
            bigrams.append((sentence[i], sentence[i+1]))
    return bigrams

bigrams = generate_bigrams(tokens)
print(bigrams)
```

### Code Explanation (Step-by-Step)

##### Step 4: Count Frequencies
- Why? Probability depends on frequency
- How? Count occurrences of:
    1. Words
    2. N-grams
    Example:
    ["I love NLP", "I love AI"]

    Counts:

    1. Unigrams 

        Word (count): I(2), love(2)
    2. Bigrams 

        (i, love) = 2, (love, nlp) = 1, (love, ai) = 1

```python
def count_freq(tokens, bigrams):
    unigram_count = defaultdict(int)
    bigram_count = defaultdict(int)

    for sentence in tokens:
        for word in sentence:
            unigram_count[word] += 1
    
    for bg in bigrams:
        bigram_count[bg] +=1
    
    return unigram_count, bigram_count

uni_count, bi_count = count_freq(tokens, bigrams)
print("Unigrams:", dict(uni_count))
print("Bigrams:", dict(bi_count))
```

- `for sentence in tokens:` → loop to iterate.
- `for word in sentence:` → loop to iterate.
- `unigram_count[word] += 1` → assigns value.
- `for bg in bigrams:` → loop to iterate.

##### Step 5: Calculate Probabilities
- Bigram Formula:

    P(wi | wi-1) = count(wi-1, wi)/ count(wi-1)

    1. Example:

        - P(love/i) = 2/2 = 1.0
        - P(nlp/love) = 1/2 = 0.5

```python
def calculate_prob(uni_count, bi_count):
    prob = {}

    for (w1,w2), count in bi_count.items():
        prob[(w1,w2)] = count/ uni_count[w1]
    return prob

probabilities = calculate_prob(uni_count, bi_count)
print(probabilities)
```

### Code Explanation (Step-by-Step)

- `def calculate_prob(uni_count, bi_count):` → defines a function.
- `prob = {}` → assigns value.

- `prob[(w1,w2)] = count/ uni_count[w1]` → assigns value.
- `return prob` → returns result.
- `probabilities = calculate_prob(uni_count, bi_count)` → assigns value.
- `print(probabilities)` → prints output.


##### Step 6: Sentence Generation
- Why? To simulate language generation
- How? 
    1. Start with a word
    2. Use probabilities to choose next word
    3. Continue
    
    Example:
    Start: "i" -> Next → "love" (prob=1) -> Next → "nlp" or "ai" (prob=0.5 each) 

```python
def sentence_generation(start_word, prob, length = 5):

    sentence = [start_word]

    for _ in range(length):
        candidates = []

        for(w1, w2), p in prob.items():
            if w1 == sentence[-1]:
                candidates.append((w2,p))
        
        if not candidates:
            break

        words = [w for w, _ in candidates]
        probs = [p for _, p in candidates]

        next_word = random.choices(words, probs)[0]
        sentence.append(next_word)
    return " ".join(sentence)

autocomplete = sentence_generation("i", probabilities)
print(autocomplete)
```

##### Step 7: Evaluation Using Perplexity
- Define: Perplexity = P(w1,w2,w3...wn)^−1/n
- Why?
    - Accuracy doesn't work because multiple correct answers exist
    - Perplexity measures uncertainty, Lower = better
- Interpretation
    1. Perplexity (Low) = Model confident
    2. Perplexity (High) = Model confused

```python
import math

def perplexity(test_sentence, prob):
    words = test_sentence.lower().split()
    n = len(words)
    p =1

    for i in range (1,n):
        if(words[i-1], words[i]) in prob:
            p *= prob[(words[i-1], words[i])]
        else:
            p *= 0.0001  # smoothing
    return pow(p, -1/n)

evaluation = perplexity(autocomplete, probabilities)
print(evaluation)
```
